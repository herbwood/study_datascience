Simple Linear Regression
---------

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 + rng.randn(50)
plt.scatter(x, y)
```

1. Selecting Model Class

```python
from sklearn.linear_model import LinearRegression
```

2. Choosing Model hyperparameters

```python
model = LinearRegression(fit_intercept = True)
model
```

3. Allocate feature matrix and target array

```python
X = x[:, np.newaxis]
X.shape
```

4. Fit the Model to the Data

```python
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

5. Predict the label of unknown datas

```python
xfit = np.linspace(-1, 11)
```

```python
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
plt.scatter(x, y)
plt.plot(xfit, yfit)
```

Supervised Learning Example : Iris Classification
    ----------------

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
X_iris = iris.drop('species', axis = 1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state = 1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. Selecting Model Class
model = GaussianNB()                       # 2. Model Instanciation
model.fit(Xtrain, ytrain)                  # 3. Fit Model to datas 
y_model = model.predict(Xtest)             # 4. Predict new datas 
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

Unsupervised Learning : Iris Dimension
    --------------

```python
from sklearn.decomposition import PCA  # 1. Selecting Model Class
model = PCA(n_components = 2)  # 2. Model Instanciation into hyperparameters
model.fit(X_iris)  # 3. Fit Model to datas 
X_2D = model.transform(X_iris)  # 4. Transform data into 2D
```

```python
iris['PCA1'] = X_2D[:, 0]
iris['PCA2'] = X_2D[:, 1]
sns.lmplot("PCA1", "PCA2", hue = 'species', data = iris, fit_reg = False)
```

Unsupervised Learning : Iris Clustering
    --------------

```python
from sklearn.mixture import GMM  # 1. Selecting Model Class
model = GMM(n_components = 3,  
           covariance_type = 'full')  # 2. Model Instanciation into hyperparameters
model.fit(X_iris)  # 3. Fit Model to datas 
y_gmm = model.predict(X_iris)  # Deciding Clustering Lable
```

```python
iris['cluster'] = y_gmm
sns.lmplot("PCA1", "PCA2", data = iris, hue = 'species',
          col = 'cluster', fit_reg = False)
```

Application : Examining Hand Written Numbers
    ---------------

Load Digit data and Visualization

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
import matplotlib.pyplot as plt

fig,axes = plt.subplots(10, 10, figsize = (8, 8),
                       subplot_kw = {'xticks' : [], 'yticks' : []},
                       gridspec_kw = dict(hspace = 0.1, wspace = 0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap = 'binary', interpolation = 'nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
           transform = ax.transAxes, color = 'green')
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

unsupervised learning : dimensionality reduction

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components = 2)
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:, 0], data_projected[:, 1], c = digits.target,
           edgecolor = 'none', alpha = 0.5,
           cmap = plt.cm.get_cmap('viridis', 10))
plt.colorbar(label = 'didit label', ticks = range(10))
plt.clim(-0.5, 9.5)
```

Digit Classification

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state = 0)
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain)
y_model = model.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat = confusion_matrix(ytest, y_model)

sns.heatmap(mat, square = True, annot = True, cbar = False)
plt.xlabel('predited value')
plt.ylabel('true value')
```

```python
fig, axes = plt.subplots(10, 10, figsize = (8, 8),
                        subplot_kw = {'xticks' : [], 'yticks' : []},
                        gridspec_kw = dict(hspace = 0.1, wspace = 0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap = 'binary', interpolation = 'nearest')
    ax.text(0.05, 0.05, str(y_model[i]),
           transform = ax.transAxes,
           color = 'green' if (ytest[i] == y_model[i]) else 'red')
```
