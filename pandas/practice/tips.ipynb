```python
# 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2, 3
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)
tips.head()
```

```python
# 4
# tips.drop(columns = ['Unnamed: 0'], inplace = True)
del tips['Unnamed: 0']
tips.head()
```

```python
# 5
# tips['total_bill'].plot(kind = 'hist')
import seaborn as sns

ttbill = sns.distplot(tips.total_bill)
ttbill.set(xlabel = 'Value', ylabel = 'Frequency', title = 'Total Bill')
sns.despine()
```

```python
# 6
# tips.plot.scatter(x = 'total_bill', y = 'tip')
sns.jointplot(x='total_bill', y = 'tip', data= tips)
```

```python
# 7
# from pandas.plotting import scatter_matrix

# attributes = ['total_bill', 'tip', 'size']
# scatter_matrix(tips[attributes], figsize=(12, 8))
sns.pairplot(tips)
```

```python
# 8
# tips2 = tips.groupby('day').sum()['total_bill']
# tips2.plot(kind='bar')
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
# 9
# tips.plot.scatter(x = 'tip', y = 'day', c = 'sex')
sns.stripplot(x='tip', y='day', hue='sex', data=tips, jitter=True)
```

```python
# 10
# tips.boxplot(by = ['time'], column = ['total_bill'])
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
```

```python
# 11
# tips.groupby('time').sum()['tip'].plot(kind = 'bar')
sns.set(style='ticks')

g = sns.FacetGrid(tips, col = 'time')
g.map(plt.hist, 'tip')
```

```python
# 12
# plt.subplot(121)
# tips.plot.scatter(x = 'total_bill', y = 'tip', c = 'smoker')
g = sns.FacetGrid(tips, col = 'sex', hue = 'smoker')
g.map(plt.scatter, 'total_bill', 'tip', alpha=.7)

g.add_legend()
```
