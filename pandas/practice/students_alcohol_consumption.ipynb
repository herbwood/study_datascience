```python
# 1 
import pandas as pd
import numpy as np
```

```python
# 2, # 3
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'

df = pd.read_csv(url)
df.head()
```

```python
# 4
df2 = df.loc[:, 'school' : 'guardian']
df2.head()
```

```python
# 5
cap = lambda x : str(x).capitalize()
```

```python
# 6
# df2[['Mjob', 'Fjob']].apply(cap)
df2['Mjob'].apply(cap)
df2['Fjob'].apply(cap)
```

```python
# 7
df2.tail()
```

```python
# 8
df['Mjob'] = df['Mjob'].apply(cap)
df['Fjob'] = df['Fjob'].apply(cap)
df.head()
```

```python
# 9
def majority(x):
    if x > 17:
        return True
    else:
        return False
    
df['legal_drinker'] = df['age'].apply(majority)
df.head()
```

```python
# 10
def times10(x):
    if type(x) is int:
        return 10 * x
    return x

df2.applymap(times10).head()

# applymap함수는 DataFrame클래스의 함수이긴 하나, 
# 위의 apply함수처럼 각 row(axis=1)나 각 column(axis=0)별로 
# 작동하는 함수가 아니라, 각 요소(element)별로 작동을 한다.
```
