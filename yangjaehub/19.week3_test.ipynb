Regression by Keras
---

```python
import tensorflow as tf
from tensorflow import keras

def make_model(x_train):
    model = keras.models.Sequential([
    keras.layers.Dense(64, activation='relu', input_dim=13), # regression이기 때문에 input_dim 값으로 입력
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(1) # regression이기 때문에 예측값 1개 출력
    ])

    return model

def main():
    boston_housing = keras.datasets.boston_housing
    (x_train, y_train), (x_test, y_test) = boston_housing.load_data()
    
    mean = x_train.mean(axis=0)
    x_train -= mean

    std = x_train.std(axis=0)
    x_train /= std

    x_test -= mean
    x_test /= std

    model = make_model(x_train)
    
    model.compile(optimizer='rmsprop', loss='mean_squared_error', metrics=['mean_absolute_error'])
    
    model.fit(x_train, y_train, epochs=80, batch_size=16)

    scores = model.evaluate(x_test, y_test)
    
    if scores is None:
        print('모델이 아직 평가되지 않았습니다.\n')
    
    else:
        print('모델의 평가 성능은 %0.5f 입니다.' % scores[1])
    
    return scores

if __name__ == "__main__":
    main()
```
