```
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np

# 기본값 설정해주기
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['ytick.left'] = plt.rcParams['ytick.labelleft'] = False
plt.rcParams['figure.figsize'] = (20,10)

# original, gray에 따라 (위아래 반전, 회전, rehape하지 않은 회전 crop, blurr, sharpen)이 가능

# get face imageof pandas from misc package
panda = misc.face()

# 오리지널 이미지
plt.subplot(151)
plt.imshow(panda)
plt.title('Original')

# 위 아래 반전 이미지
plt.subplot(152)
flip_down = np.flipud(misc.face())
plt.imshow(flip_down)
plt.title('Up Down Flip')

# 135도 회전된 이미지
plt.subplot(153)
panda_rotate = ndimage.rotate(panda, 135)
plt.imshow(panda_rotate)
plt.title('Rotation')

# gray scale 적용된 이미지
plt.subplot(154)
face = misc.face(gray=True)
plt.imshow(face)
plt.title('Gray')

# gray scale 적용 + cropped
plt.subplot(155)
crop_face = face[lx // 4: - lx // 4, ly // 4: - ly // 4]
plt.imshow(crop_face)
plt.title('Gray Cropped')
```

```
# gray scale + 위아래 반전
plt.subplot(161)
flip_ud_face = np.flipud(face)
plt.imshow(flip_ud_face)
plt.title('Gray Flipped Updown')

# gray scale + 45도 회전
plt.subplot(162)
rotate_face = ndimage.rotate(face, 45)
plt.imshow(rotate_face)
plt.title('Gray Rotated')

# gray scale + 45도 회전 + reshape 하지 않음
plt.subplot(163)
rotate_face_noreshape = ndimage.rotate(face, 45, reshape=False)
plt.imshow(rotate_face_noreshape)
plt.title('Gray Rotated No Reshaped')

# gray scale + blurr 약하게
plt.subplot(164)
face2 = misc.face(gray=True)
blurred_face = ndimage.gaussian_filter(face2, sigma=3)
plt.imshow(blurred_face)
plt.title('Gray Blurred')

# gray scale + blurr 강하게
plt.subplot(165)
very_blurred = ndimage.gaussian_filter(face2, sigma=5)
plt.imshow(very_blurred)
plt.title('Gray very Blurred')

# gray scale + sharpend
plt.subplot(166)
face3 = misc.face(gray=True).astype(float)
blurred_f = ndimage.gaussian_filter(face3, 3)
filter_blurred_f = ndimage.gaussian_filter(blurred_f, 1)
alpha = 30
sharpend = blurred_f + alpha * (blurred_f - filter_blurred_f)
plt.imshow(sharpend)
plt.title('Gray sharpend')
```
