Matplotlib Cheetsheets
===========

일반적인 그래프
-------

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()

plt.rcParams['ytick.labelsize'] = 12

x = np.linspace(0, 10, 1000)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(2, sharex = True, figsize = (12, 8))
ax[0].plot(x, y)
ax[0].set(ylabel = 'y', 
          xlim = (10, 0), ylim = (1.2, -1.2),
         title = 'A Simple Plot')

ax[1].plot(x, z)
ax[1].set(xlabel = 'x')

plt.show()
```

산점도
---

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.figure(figsize = (10, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
           cmap='viridis')
plt.colorbar()
plt.show()
```

히스토그램
-----

```python
plt.style.use('fivethirtyeight')

np.random.seed(19680801)
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(437)
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(1, 2, figsize = (12, 5))
ax[0].hist(x, bins = 50)
ax[0].set(xlabel = 'Smarts', ylabel = 'probability density', title = 'Histogram')

ax[1].plot(x1, y1, 'o-')

fig.tight_layout()
plt.show()
```

한글 폰트 사용하기
-------

```python
import platform
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt

plt.rcParams['axes.unicode_minus'] = False # 마이너스 표시 오류 제거

if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    path = 'C:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system...')
```
